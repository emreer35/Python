# 5) Concatenation (Array Birlestirme)

### Tek boyut uzerine array birlestirme

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
```

```python
np.concatenate([x,y])
```

yeni bir array olsuturup onu da ekleyebilir miyim

```python
z = np.array([7,8,9])
```

```python
np.concatenate([x,y,z])
```

### iki boyutlu arraylarda birlestirme

```python
a = np.array([[1,2,3],
              [4,5,6]])
```

#### iki boyutlu array birlestirme

```python
np.concatenate([a,a])
```

#### iki boyutlu arrayi axis e gore birlestirme
* axis = 0

```python
np.concatenate([a,a],axis=0) # axis = 0 satira gore birlestirir auto degeri 0 dir
```

* axis = 1
* axis = 1 i sutuna bolerek birlestirme islemi yapti

```python
np.concatenate([a,a], axis=1)
```

# 6) Splitting (Array Ayirma)

### Bir bouytlu arraylarda bolme

```python
x = np.array([1,2,3,99,99,3,2,1])
```

```python
np.split(x,[3,5]) # 0 dan 3 e kadar 3 ten 5 e kadar gerisini de oylece koy
```

bu degerleri variable a atamak icin ne yapiyoruz

```python
a,b,c = np.split(x, [3,5])
```

```python
c
```
