# 14) Numpy ile iki bilinmeyenli denklem cozumu

* ### lineer cebir alanina gore gelistirilen **linalg**

5 * x0 + x1 = 12

x0 + 3* x1 = 10

iki bilinmeyenli bir denklemi bulabilmek icin yaptigimiz islem
* ilk basta bilinmeyenlerin kat sayilarini bir arraya aliriz ayri ayri
* sonucu farkli bir arraya aliriz
* sonrasinda linalg fonk solve methodunu kullaniriz

```python
a = np.array([[5,1], [1,3]])
b = np.array([12,10])
```

```python
solve = np.linalg.solve(a,b)
solve
```
